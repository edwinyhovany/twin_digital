# 05_DataValidation

Este notebook muestra cómo validar datos numéricos y categóricos usando el módulo `src.validation.validation`.
Se utilizan rangos numéricos y listas de valores permitidos para detectar inconsistencias.

```python
from src.data_loader.data_loader import load_parquet
from src.sampling.sample import random_sample
from src.validation.validation import validate_numeric_ranges, validate_categorical_consistency
import json

# Cargar una muestra de datos
raw = load_parquet('COSTOS_PAGOS_DIABETES_2022.parquet')
sample = random_sample(raw, n=1000)

# Definir reglas de validación (ejemplo)
numeric_rules = {
    'age': (18, 100),
    'payment_amount': (0, 1_000_000)
}
categorical_rules = {
    'sex': {'M', 'F'},
    'region': {'North', 'South', 'East', 'West'}
}

num_report = validate_numeric_ranges(sample, numeric_rules)
cat_report = validate_categorical_consistency(sample, categorical_rules)

print('Numeric validation report:')
print(json.dumps(num_report, indent=2))
print('
Categorical validation report:')
print(json.dumps(cat_report, indent=2))
```
