This notebook demonstrates loading all yearly Parquet files using the `src.data_loader` module and saving an interim version.

```python
from src.data_loader.data_loader import load_all_years
import pathlib

# Load all years of cost data
raw_df = load_all_years()

# Save a subset (first 2000 rows) to the interim folder for later steps
interim_path = pathlib.Path('data/interim') / 'raw_all_years_sample.parquet'
raw_df.head(2000).to_parquet(interim_path, engine='pyarrow')
print('Saved interim sample to', interim_path)
```
